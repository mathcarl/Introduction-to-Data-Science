# Loops
 Loops are a reapeated operation that is executed. This can be perfomed by for and while loops.

```
for i in range(0,8): # Will repeat the sequence until reaching 7.
  print(i)
```

```
for x in range(3,10,2): # This code will give the command to (start,stop,steps). 
  print(x) # This command will only print the steps mentioned in the function.
```

While loops will keep executing a code based on a condition and will only stop the loop until the condition is False based on boolean value. 

```
# Will add 1 every loop until count is <5
count=0
while count < 5: # count is less than 5
  print(count)
  count +=1  # count + 1
```

Break and Continue are used when the while loop becomes False. This will give the order to continue to another code or exit the loop. 

```
count=0
while True:
  print(count)
  count+=1
  if count >= 5: # Will count to 5 and then will break and stop the loop
    break 
  

for x in range(10): # Print out only odd number 1,3,5,7,9
  if x%2==0: # Check if x is an even number using modulo
    continue
  print(x)

```

Else is used when (for) and (while) loops fail and we want the code to follow another condition named else. 

```
count=0
while (count<5):
  print(count)
  count +=1
else:
  print(f'count value reached {count}')
```

#Functions
 A function is a block that allows you to perform certain parameter or argument and be later reused. They let you break down a task to a more simple and readable code. Some rules that we follow to define a function are:


1.   We start the function by first defining the name and parameter to be used(def name():).
2.  Followed by the documentation(optional)enclosed by """Example""".
3.  Now we could define our body of the function after the colon(:)
4.  This function is exited by the command return(). 



```
#Building a function
def add(x): # Define the function name and parameter
   """Will add two to x"""
   y= x+2 # Body of the function
   print(x,"if you add two ",y) # Will print the parameter of x and the result y
   return (y) # Result of y
   
```

```
add(100) # Recalling the function
```

```
# Below is the exercise for functions 
def list_benefits(): # Function will return the list of strings
    # Will return each individual string in the list
    return ("More organized code", "More readable code", "Easier code reuse", 
            "Allowing programmers to share and connect code together")

# This function will allow you to add a string to the ending msg
def build_sentence(benefit): #Creates a new function to attach a string
    return f"{benefit} is a benefit of functions!" 

# Function will allow the previous two functions to be put together
def name_the_benefits_of_functions(): # Name a new function
    list_of_benefits = list_benefits()# Change the name of the variable
    for benefit in list_of_benefits: # Calls the list to benefits the parameter of the second function
        print(build_sentence(benefit)) # Use the second function to put together the new sentence

name_the_benefits_of_functions()
```

#Pandas
Pandas is a python library that help us analyse data that depends on other libraries like numpy and matplotlib. Some of the ways we can use pandas is to load and save files, viewing data, and manipulate data to name a few.

```
#Pandas allow us to create a dataframe from a dictionary, list, or numpy array.
# Here is the book example from a dictionary, but it would be the similar to a list or array
# Below we can see a dictionary with keys and values.
dict = {"country": ["Brazil", "Russia", "India", "China", "South Africa"],
       "capital": ["Brasilia", "Moscow", "New Dehli", "Beijing", "Pretoria"],
       "area": [8.516, 17.10, 3.286, 9.597, 1.221],
       "population": [200.4, 143.5, 1252, 1357, 52.98] }

import pandas as pd
brics = pd.DataFrame(dict)
print(brics)
```

```
# .Index will allow us to change the key value from numerical to a selected two letter representation of the country.
brics.index = ["BR","RU","IN","CH","SA"] #  Assign the values to the dataframe key instead of [0,1,2,3,4].
print(brics)
```

We can also import files to pandas and  be able to manipulated then. This files are csv, excel, and others compatibles files to pandas. Below we can see an example were a csv file was converted to a dataframe.

csv file.png

```
# Indexing data frames
# From the previous example we indexed few of the columns in the data.
print(brics['country']) # Will ouput a data series
print(brics[['capital']])# Will output a dataframe
print(brics['population'])


```

```
# We can also print rows in the data for example
print(brics[0:3]) # We can use slicing to output the data
```

Using iloc and loc allows us to select data in pandas. 

```
# Selecting data using loc
# From the previous example we select Brazil and Russia
import pandas as pd
# for loc we select  the row index and the column header
print(brics.loc["BR","country"]) # row BR and column country
print(brics.loc['RU','country'])
```

```
# Selecting data using iloc
# iloc is integer based and we find a position in the data fram using column and row number.

print(brics.iloc[0,0]) # as mentioned we can find a location by row and column number.
print(brics.iloc[1,2])
```
